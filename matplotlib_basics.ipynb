Scatter plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
```

```python
plt.scatter(df['Age'], df['Fare'], c=df['Pclass'])
plt.xlabel('Age')
plt.ylabel('Fare')
```
