Import Numpy and Pandas

```python
import numpy as np
import pandas as pd
```

Load a pandas dataframe

```python
df = pd.read_csv('https://sololearn.com/uploads/files/titanic.csv')
df.head()
```

Convert a pandas series to a numpy array

```python
df['Fare'].values
```

Convert a pandas dataframe to a numpy array (2 dimensional)

```python
print(df[['Pclass', 'Fare', 'Age']].values)
```

To determine the size of our numpy array we use the shape attribute.

```python
arr = df[['Pclass', 'Fare', 'Age']].values
arr.shape
```

Selecting a single element form the array

```python
arr[0, 1]
```

Selecting a single row

```python
arr[0]
```

Select a single column

```python
arr[:, 2]
```

Masking: selecting all the rows that fulfull a certain criteria.  
First we create a mask of boolean values and then we use the mask to select the values that fulfull the criteria defined by the mask.

```python
mask = arr[:, 2] < 18    # Age less than 18 years
arr[mask]
```

We don't need to explicitly define a mask, we can also do this in a single line.

```python
arr[arr[:, 2] < 18]
```

Another way to use the mask is to get number of children on the titanic

```python
(arr[:, 2] > 18).sum()  # returns the number of true values. 
```
